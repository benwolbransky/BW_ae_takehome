```python
# 1) What is the total counts of offer activations for each customer?

# Simple sql query for number of unique activations per customer, ordered from most to fewest.
c.query('SELECT customer_id, COUNT(DISTINCT id) from customer_offers \
         GROUP BY 1 \
         ORDER BY 2 desc')
```

```python
# 2) Provide a list of customers that who haven't activated an offer in the last couple months?

# It looks like the data I have spans only a week, so I'm going to answer this as "last couple days"
# I built a python function to check last n days (assuming today is March, 24, 2021, the most recent date in data)

def non_active_customers(lookback = 3):
    if lookback >= 7:
        print("the dataset doesn't go back that far.")
        return
    non_active_query = f'''
    with all_customers as (SELECT distinct customer_id FROM customer_offers),
    last_days as (SELECT customer_id from customer_offers where ACTIVATED >= DATE('2021-03-24', '-{lookback} days'))

    SELECT customer_id from all_customers
    EXCEPT
    SELECT customer_id from last_days
    '''
    list_of_custs = c.query(non_active_query)
    print(f"There are {len(list_of_custs)} customers who haven't activated an offer in {lookback} days")
    return list_of_custs

non_active_customers(6)
```

```python
# 3) What is the conversion rate of activated to complete for each customer?

# I made a CTE here because when I was taking the count of verifications it was counting NULLS.
#   At least, I think that's what was happening. I was getting strange answers
#   and felt making a CTE was the easiest way to keep moving along.
#   I also used cast as REAL to get the decimals so you could do 
#   verifcations / activations to get a percentage.

c.query('''
    WITH counts as (SELECT customer_id, 
                    COUNT(activated) as activations,
                    COUNT(verified) as verifications
                    FROM customer_offers
                    GROUP BY 1)

    SELECT customer_id, CAST(sum(activations) as REAL)/CAST(sum(verifications) as REAL) as activate_to_conversion_rate from COUNTS
    GROUP BY 1
    ORDER BY 2 DESC
        ''')
```

```python
# 4) What is the total amount of redemption for each customer?

# I had to join the customer offer redemptions table to customer offers as 
# customer ID isn't in the redemptions table. It was a simple enough join.
# I also rounded the sum of amounts to remove trailing numbers. 

c.query('''
    SELECT cus.customer_id, ROUND(sum(red.offer_amount),2) as total_amount
    FROM customer_offers cus
    JOIN customer_offer_redemptions red
    ON red.customer_offer_id = cus.id
    GROUP BY 1
    ORDER BY 2 DESC
        ''')
```
